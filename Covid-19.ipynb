## Q2. Remove all the Records where Confirmed Cases is less than 10.

```python
data.head()
```

```python
data[data['Confirmed'] < 10] # here we can see the all confirmed cases which is less than 10 .
```

```python
indices_to_drop = data[data['Confirmed'] < 10].index   # here we have used drop method to remove the value 
# which is less than 10.
data = data.drop(indices_to_drop)
```

```python
data.head(20)
```

## In which region maximum number of cases are recorded ?

```python
grouped_data = data.groupby('Region')['Confirmed'].sum()  # Group by 'Region' and sum the 'Confirmed' cases
max_region = grouped_data.idxmax()    # Find the region with the maximum number of confirmed cases
max_cases = grouped_data.max()
print(max_cases, max_region)
```

```python
data.groupby('Region')['Confirmed'].sum().sort_values(ascending = False).head(10)  # this  is the another way 
#to find  the max cases in the confirmed cases and max cases in the region 
```

## Find the minimum number of Death Cases ?

```python
data.groupby('Region')['Deaths'].sum().sort_values(ascending=True).head(20)
```
